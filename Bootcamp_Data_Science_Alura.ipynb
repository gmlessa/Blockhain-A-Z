###Desafio 2: faça a mesma análise para o mês mais recente

```python
##Quando fazemos dados.mean(), ele utiliza somente as colunas que tem todos as linhas preenchidas, 
##portanto foi utilizado essa forma para utilizar somente os meses em que todos os dados estão preenchidos
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis[:5]
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
##O mês mais recente é o -2 porque o -1 é o Total
mes_mais_recente = usaveis.columns[-2]

ax = dados.plot(x="Unidade da Federação", y = f"{mes_mais_recente}", kind = "bar", figsize = (10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor gasto total com procedimentos hospitalares do SUS por Unidade da Federação desde 2008")
plt.show()
```

###Desafio 1: Reposicionar a legenda. Dentro? Fora? Onde?

```python
ax = usaveis_sem_total.T.plot(figsize=(10,6))
#ncol muda o número de colunas na legenda e loc muda a localização da legenda, bbox_to_anchor permite fazer um ajuste da posição de forma mais livre
plt.legend(bbox_to_anchor=(1.02, 1.02), loc="upper left", ncol=2, title="Unidades da Federação")
plt.show()
```

###Desafio 2: Selecionar 5 estados de sua preferência

```python
for i in usaveis_sem_total.index:
  if (i[0] == "1"):
    usaveis_sem_total["Região"].append("Norte")
  if (i[0] == "2"):
    usaveis_sem_total["Região"].append("Nordeste")
  if (i[0] == "3"):
    usaveis_sem_total["Região"].append("Sudeste")
  if (i[0] == "4"):
    usaveis_sem_total["Região"].append("Sul")
  if (i[0] == "5"):
    usaveis_sem_total["Região"].append("Centro-Oeste")

  usaveis_sem_total["Região"]
```
